```python
import pandas as pd
```

```python
df = pd.read_csv("data/synthetic_financial_data.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
#payment and debt dataframe
selected_cat = ['PAYMENT','DEBT']
mask = df['type'].isin(selected_cat)
df_pd = df[mask]
#droping fraud columns
df_pd = df_pd.drop(['isFraud','isFlaggedFraud'], axis=1)
```

```python
df_pd
```

```python
# Transforming step into week, day and hour
df_pd['week'] = (df_pd['step'] - 1) // (24 *7) + 1
df_pd['day'] = (df_pd['step'] - 1) // 24 + 1
df_pd['hour'] = (df_pd['step'] - 1) % 24
```

```python
df_pd
```

```python
df_pd.to_csv('data/transformed_data.csv', index=False)
```
